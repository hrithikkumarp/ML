```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.svm import SVC
```

```python
file_path = '/content/drive/MyDrive/world_population.csv'
```

```python
data=pd.read_csv(file_path)
data.head()
```

```python
data.isnull().sum()
```

```python
data['Country/Territory'].nunique()
```

```python
data.shape
```

```python
x1=data['Continent']
y1=data['Growth Rate']
```

```python
plt.bar(x1, y1, label="bar", color='blue')
plt.xlabel("Continent")
plt.ylabel("Rate of Growth")
plt.title("Rate of Growth based on Continents")
plt.legend()
plt.show()
```

```python
sort_Population = data.sort_values(by='2022 Population', ascending=False)
top_populated=sort_Population.head(10)
top_populated
```

```python
plt.scatter(top_populated['2022 Population'],top_populated['Country/Territory'],marker='*',color='red')
plt.title('Top 10 Populated country')
plt.show()
```

```python
least_populated=sort_Population.tail(10)
least_populated
```

```python
plt.plot(least_populated['2022 Population'],least_populated['Country/Territory'], label='Bar', color='green', marker='+',markersize=10)
plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Least-Populated')
plt.legend()
plt.show()
```
