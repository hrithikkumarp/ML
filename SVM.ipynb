```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.svm import SVC
```

```python
data=pd.read_csv("https://raw.githubusercontent.com/nunnarilabs/ml/master/logistic_regression/insurance_data.csv")
data.head()
```

```python
data.describe()
```

```python
data.nunique()
```

```python
X=data.drop('bought_insurance',axis=1)
Y=data.bought_insurance
Y
```

```python
from sklearn import linear_model as LM
```

```python
model=LM.LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,train_size=0.8)
print(len(X))
print("Train size",len(X_train))
print("Test size",len(X_test))
```

```python
model.fit(X_train,Y_train)
```

```python
print(model.coef_)
print(model.intercept_)
```

```python
Y_pred=model.predict(X_test)
Y_pred
```

```python
print("Test Score:",model.score(X_test,Y_test))
print("Train Score:",model.score(X_train,Y_train))
```

```python
titanic=pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Titanic/train.csv")
```

```python
titanic.head()
```

```python
titanic.drop(['PassengerId','Ticket','Name'],axis=1,inplace=True)
titanic.head()
```

```python
titanic.isnull().sum()
```

```python
titanic[titanic.Embarked.isnull()]
```

```python
titanic.drop(['Cabin'],axis=1,inplace=True)
titanic["Age"] = titanic["Age"].fillna(titanic["Age"].median())
titanic["Embarked"] = titanic["Embarked"].fillna("C")
titanic.Sex=titanic.Sex.replace(['male','female'],[0,1])
titanic.Embarked=titanic.Embarked.replace(['S','C','Q'],[0,1,2])
```

```python
titanic.dropna(inplace=True)
```

```python
test=pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Titanic/test.csv")
test.head()
```

```python
test.drop(['Cabin'],axis=1,inplace=True)
test["Age"] = test["Age"].fillna(test["Age"].median())
test["Embarked"] = test["Embarked"].fillna("C")
test.Sex=test.Sex.replace(['male','female'],[0,1])
test.Embarked=test.Embarked.replace(['S','C','Q'],[0,1,2])
```

```python
test.dropna(inplace=True)
```

```python
features=["Pclass","Sex","Age","SibSp","Parch","Fare","Embarked"]
```

```python
c=0.7
clf = svm.SVC(kernel='linear',C=c).fit(titanic[features],titanic["Survived"])
```

```python
prediction = clf.predict(test[features])
```
