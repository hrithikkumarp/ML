```python
avg_loss_auto=auto['normalized-losses'].astype("float").mean(axis=0)
avg_loss_auto
```

```python
auto['normalized-losses'].replace(np.nan,avg_loss_auto,inplace=True)
auto.isna().sum()
```

```python
auto['bore'].replace(np.nan,auto['bore'].astype("float").mean(axis=0),inplace=True)
auto['stroke'].replace(np.nan,auto['stroke'].astype("float").mean(axis=0),inplace=True)
auto['horsepower'].replace(np.nan,auto['horsepower'].astype("float").mean(axis=0),inplace=True)
auto['peak-rpm'].replace(np.nan,auto['peak-rpm'].astype("float").mean(axis=0),inplace=True)
auto['price'].replace(np.nan,auto['price'].astype("float").mean(axis=0),inplace=True)
```

```python
auto.dropna(subset=['price'],inplace=True)
auto.reset_index(drop=True,inplace=True``)
```

```python
auto['normalized-losses']=auto['normalized-losses'].astype('int')
```

```python
#z-score operation
from scipy import stats
hp_zscore=np.abs(stats.zscore(auto['normalized-losses']))
hp_zscore
```

```python
threshold=hp_zscore.max()
```

```python
index_val=np.where(hp_zscore >= threshold)
auto.iloc[index_val]['normalized-losses']
```

```python
new_auto=auto.drop(index=auto.iloc[index_val]['normalized-losses'].index)
new_auto
```

```python
new_auto.reset_index(drop=True,inplace=True)
```
